```python
import pandas as pd
import numpy as np
import xgboost as xgb

df = pd.read_csv(r"D:\NATPAC\Datasets\combined.csv")
df.drop(df.columns[0], axis=1, inplace=True)

X = df.iloc[:, :-1].values
y = df.iloc[:, -1].values

model_xgb = xgb.XGBRegressor()
model_xgb.fit(X, y)

forecast_2023 = model_xgb.predict(X)

print(forecast_2023)
```

```python
import matplotlib.pyplot as plt
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
month_ticks = np.arange(0, 366, 30.5)
days = range(366)
plt.figure(figsize=(15, 8))
plt.plot(days, forecast_2023, marker='')
plt.xlabel('Months')
plt.ylabel('Quantity')
plt.title('Predicted Data for 2023')
plt.xticks(month_ticks, months)
plt.xlim(0, 366)
plt.show()
```

```python
import pandas as pd

df2 = pd.read_csv(r"D:\NATPAC\Datasets\combined.csv")

new_df = pd.DataFrame(forecast_2023, columns=["2023"])

result_df = pd.concat([df2, new_df], axis=1)

result_df.to_csv(r"D:\NATPAC\Datasets\XGBoost\2023.csv", index=False)

print(result_df)
```
