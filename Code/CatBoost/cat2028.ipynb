```python
import pandas as pd
import numpy as np
from catboost import CatBoostRegressor

df = pd.read_csv(r"D:\NATPAC-Alosh\Datasets\CatBoost\2027.csv")
df.drop(df.columns[0], axis=1, inplace=True)

X = df.iloc[:, :-1].values
y = df.iloc[:, -1].values

model_cat = CatBoostRegressor(n_estimators=50000)
model_cat.fit(X, y)

forecast_2028_cat = model_cat.predict(X)

print(forecast_2028_cat)
```

```python
import matplotlib.pyplot as plt
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
month_ticks = np.arange(0, 366, 30.5)
days = range(366)
plt.figure(figsize=(15, 8))
plt.plot(days, forecast_2028_cat, marker='')
plt.xlabel('Months')
plt.ylabel('Quantity')
plt.title('Predicted Data for 2028')
plt.xticks(month_ticks, months)
plt.xlim(0, 366)
plt.show()
```

```python
import pandas as pd

df2 = pd.read_csv(r"D:\NATPAC-Alosh\Datasets\CatBoost\2027.csv")

new_df = pd.DataFrame(forecast_2028_cat, columns=["2028"])

result_df = pd.concat([df2, new_df], axis=1)

result_df.to_csv(r"D:\NATPAC-Alosh\Datasets\CatBoost\2028.csv", index=False)

print(result_df)
```
