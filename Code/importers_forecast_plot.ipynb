```python
import matplotlib.pyplot as plt
import pandas as pd
import matplotlib.dates as mdates

# Read the CSV file
# Replace 'your_data.csv' with the actual name of your CSV file
df = pd.read_csv(r"D:\NATPAC-Alosh\Datasets\Importers\Linear_impo.csv", parse_dates=['Sail Date.'])

# Select the years we want to plot (2024-2028)
years_to_plot = ['2024', '2025', '2026', '2027', '2028']

# Create the plot
plt.figure(figsize=(12, 6))

# Plot each year with a different color
for year in years_to_plot:
    plt.plot(df['Sail Date.'], df[year], label=year)

plt.title('Data for Years 2024-2028')
plt.xlabel('Date')
plt.ylabel('Reading Value')
plt.legend()
plt.grid(True)

# Format x-axis to show dates nicely
plt.gca().xaxis.set_major_locator(mdates.MonthLocator())
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%b'))

# Add a secondary x-axis for the year
# secondary_ax = plt.gca().secondary_xaxis('top')
# secondary_ax.xaxis.set_major_locator(mdates.YearLocator())
# secondary_ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y'))

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Read the CSV file
df = pd.read_csv(r"D:\NATPAC-Alosh\Datasets\Importers\Rand_impo.csv")

# Select the years we want to plot (2024-2028)
years_to_plot = ['2024', '2025', '2026', '2027', '2028']

# Create the plot
plt.figure(figsize=(15, 8))

# Colors for each year
colors = ['orange', 'green', 'purple', 'blue', 'red']

# Plot each year's data with its respective color
start = 0
for year, color in zip(years_to_plot, colors):
    end = start + len(df)
    plt.plot(range(start, end), df[year], color=color, label=year, linewidth=0.8)
    start = end

plt.title('Random Forest', fontsize=16)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Quantity (in 10,000 TONS)', fontsize=12)
plt.legend(loc='upper right')
plt.grid(True, which='both', linestyle='--', linewidth=0.5)

# Format x-axis to show years
year_positions = np.arange(0, len(df) * len(years_to_plot), len(df))
plt.xticks(year_positions, years_to_plot)

# Use a scientific notation for y-axis
plt.ticklabel_format(style='sci', axis='y', scilimits=(0,0))

plt.tight_layout()
plt.show()
```
