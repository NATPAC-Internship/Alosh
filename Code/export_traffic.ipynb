```python
import pandas as pd

for year in range(2015, 2024):
    # Read the dataset for the selected year
    file_path = f"D:/NATPAC/Datasets/Exporters/EXP {year}.xlsx"
    df = pd.read_excel(file_path)

    # Create a new DataFrame to store the processed data
    new_df = pd.DataFrame(columns=['Exporter', 'Product', 'QTY.(KGS)'])
    data = []
    # Iterate through the rows and identify the "TOTAL" rows
    for i, row in df.iterrows():
        if row['Exporter'] == 'TOTAL                                   ':
            # Get the exporter name from the previous row
            prev_row = df.iloc[i-1]
            exporter = prev_row['Exporter']
            product = prev_row['Product']
            quantity = row['QTY.(KGS)']
            
            # Add the data to the new DataFrame
            data.append({'Exporter': exporter, 'Product': product, 'QTY.(KGS)': quantity})

    new_df = pd.DataFrame(data)
    print(new_df)
    new_df.to_csv(f"D:/NATPAC/Datasets/Exporters Totalled/EXP {year}.csv", index=False)
```

```python
import pandas as pd

exporters_df = pd.read_csv(r"D:\NATPAC\Datasets\exporters_with_pincode.csv")
for year in range(2015, 2024):
    exporters_df[str(year)] = 0
for year in range(2015, 2024):
    yearly_data = pd.read_csv(f"D:/NATPAC/Datasets/Exporters Totalled/EXP {year}.csv")
    
    for _, exporter_row in exporters_df.iterrows():
        exporter_name = exporter_row["Exporter Name"]
        
        # Check if the exporter exists in the yearly data
        matched_rows = yearly_data[yearly_data["Exporter"] == exporter_name]
        if not matched_rows.empty:
            # Get the total quantity for the exporter in the current year
            total_quantity = matched_rows["QTY.(KGS)"].sum()
            exporters_df.at[_, str(year)] = total_quantity
exporters_df.to_csv("D:/NATPAC/Datasets/exporters_with_yearly_data.csv", index=False)
```

```python
import pandas as pd

exporters_df = pd.read_csv("D:/NATPAC/Datasets/exporters_with_yearly_data.csv")
pincode_df = pd.read_csv("D:/NATPAC/Datasets/all-india-pincode-directory.csv")

exporters_df['Taluk'] = ''
exporters_df['District'] = ''

for _, exporter_row in exporters_df.iterrows():
    pincode = str(exporter_row["pin"])
    
    # Find the matching row in the pincode_df
    matching_rows = pincode_df[pincode_df["pincode"] == int(pincode)]
    
    if not matching_rows.empty:
        # Get the Taluk and District values from the matching row
        taluk = matching_rows["Taluk"].values[0]
        district = matching_rows["Districtname"].values[0]
        
        # Update the exporters_df with the Taluk and District values
        exporters_df.at[_, 'Taluk'] = taluk
        exporters_df.at[_, 'District'] = district

exporters_df.to_csv("D:/NATPAC/Datasets/exporters_with_yearly_data_and_location.csv", index=False)
```
